# as *counter_number* and *reading_remarque* are both numerical data we can use either median or mode to replace missing values

# mean is excluded as it returns float result which will not help us in our case

```python
df['counter_number'].fillna(df['counter_number'].median(),inplace=True)
```

```python
df['reading_remarque']=df['reading_remarque'].fillna(df['reading_remarque'].median())
```
